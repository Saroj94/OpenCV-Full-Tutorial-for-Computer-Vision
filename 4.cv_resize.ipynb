```python
import cv2
```

```python
##resize the image
##load the image
org_img=cv2.imread("Temi-Tea Namchi.png")

##check the file
if org_img is None:
    print("Error: Image not found.")
else:
    print("Image is successfully loaded.")
    resized_img=cv2.resize(org_img, (300, 300),interpolation=cv2.INTER_AREA) # interpolation=cv2.INTER_AREA
    cv2.imshow("Original Image", org_img)
    cv2.imshow("Resized Image", resized_img)

    ##save the resized image as well
    cv2.imwrite("resized_img.png", resized_img)
```
