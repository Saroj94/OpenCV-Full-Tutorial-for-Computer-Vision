## **Color Conversion**

```python
## libraries
import cv2
```

```python
##load data
img = cv2.imread("AI.jpg")

##image 
if img is not None:
    gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Grey Image Title",gray_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Could not load the image")
```
