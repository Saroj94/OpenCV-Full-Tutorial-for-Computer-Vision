```python
##load libraries
import cv2
```

```python
##load images
img = cv2.imread("Temi-Tea Namchi.png")
```

## **Image with Lines**

```python
##image with lines

if img is None:
    print("Error: Image not found.")
else:
    ##attributes
    pnt1=(50,100) ## 50 pixels to X-Axis, 100 pixels to Y-Axis
    pnt2=(300,100) ## 300 pixels to X-Axis, 100 pixels to Y-axis
    color=(0,0,255)
    thick=4

    ##drawing line in the image
    img_lined=cv2.line(img,pnt1,pnt2,color=color, thickness=thick)

    ##display
    cv2.imshow("Draw a line in image", img_lined)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## **Image with Rectangle Box**

```python
if img is None:
    print("Error: Image is not found.")
else:
    ##corner points
    cor_pnt1=(100,100)
    cor_pnt2=(200,200)

    ##colors
    color_rec = (0,0,255)

    ##thickness
    thick = 4

    ##apply on image
    rec_img = cv2.rectangle(img, cor_pnt1,cor_pnt2, color=color_rec, thickness=thick)

    ##display image
    cv2.imshow("Drawing rectangle box on image", rec_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## **Image with Circle**

```python
if img is None:
    print("Error: Image not found")
else:
    ##center
    center_pnt=(100,150)
    ##radius
    radius=50 ##50 pixels

    color_cir=(255,0,0)

    ##thickness 
    thick=3

    ##apply on image
    img_cir = cv2.circle(img, center=center_pnt, radius=radius,thickness=thick,color=color_cir)

    ##display
    cv2.imshow("Drawing circle on image",img_cir)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
