```python
##load libraries
import cv2
```

```python
##load images
img = cv2.imread("Temi-Tea Namchi.png")

if img is None:
    print("Error: Image not found.")
else:
    ##attributes
    pnt1=(50,100)
    pnt2=(300,100)
    color=(0,0,255)
    thick=4

    ##drawing line in the image
    img_lined=cv2.line(img,pnt1,pnt2,color=color, thickness=thick)

    ##display
    cv2.imshow("Draw a line in image", img_lined)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
