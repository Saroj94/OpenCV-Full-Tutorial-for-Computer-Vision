## Assignment-1
1. Take location of image from user
2. Display that image
3. Rename the saved file by taking it from users
4. Save that image in grayscale
5. Print the message saying "File name is saved succesfully"

```python
import cv2
```

```python
##image path/location
img_path="Temi-Tea Namchi.png"
```

```python
##Assignment
img_loc=input("Enter the Image Location: ")
if img_loc is not None:
    ##read image
    cv_img=cv2.imread(img_loc)

    ##Taking new name from user to rename the file to save it
    new_name=input("Enter the new Image name: ")

    ##convert the color of the image into a grayscale
    gray_img=cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)

    ##saving the image with new name
    output_img=cv2.imwrite(new_name,gray_img)
    if output_img:
        print(f"\nImage is successfully saved by the name: {new_name}")
    else:
        print("Image not found to saved or new name is not provided.")
        
    ##displaying image
    print("Displaying the Color image")
    cv2.imshow("Image",cv_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

else:
    print("Image not found to load")
```
