```python
##load library
import cv2
```

```python
##load image
img = cv2.imread("Temi-Tea Namchi.png")

if img is None:
    print("Error: Image not loaded")

else:
    ##get the image width and height
    (w,h) = img.shape[:2]
    ##center of the rectangular image
    center = (w//2, h//2)
    ##angle to rotated
    angle = 90
    ##scale of the image
    scale=1.0
    ##create formula
    M = cv2.getRotationMatrix2D(center, angle, scale)
    ##apply the formula into the original image
    rotated_img = cv2.warpAffine(img,M,(w,h))

    ##display the original and roted image
    cv2.imshow("Original image", img)
    cv2.imshow("Rotated Image", rotated_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
