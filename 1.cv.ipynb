## Phase -1: Basics of Image Handling
**Basic Terms of Image** : When we as a ML engineer see image this terms needs to fully proficient.
- Image: small grid/matrix
- Pixel: numbers
- Width and Height: total number of pixel on vertically and horizotally
- Color Channel: 
    - Greyscale: 1
    - RGB or BGR: 3 
        - Pure Blue:[255, 0, 0]
        - Pure Green:[0, 255, 0]
        - Pure Red: [0, 0, 255]
- Image Format: jpg, png, bmp etc.
    - .jpg: small and losely compressed
    - .png: high quality image
    - .bmp: uncompressed images, used in hardware apps.
    - .tiff: Super high resolution, used in scientific work
    

```python
## opencv library
import cv2
import matplotlib.pyplot as plt
```

## **Read Image**

```python
##open and read image
##flags has three values: 1-color image, 0-greyscale, -1 - unchanged
file_path="/Users/sarojrai/Desktop/Computer Vision/OpenCV for Computer Vision/AI.jpg"
AI_image=cv2.imread(file_path, flags=1)

if AI_image is None:
    print("Error: Image not found")
else:
    print("Image Successfully loaded")
```

## **Save Image after manipulation**

```python
##saving the image
if AI_image is not None:
    output_img=cv2.imwrite("New_output_Image.png",AI_image)
    if output_img:
        print("Edited image is successfully saved as output_image.png")
    else:
        print("Error: failed to save the image")
else:
    print("Error: could load image.")
```
