## **Dimension of Image**

```python
import cv2 
```

```python
##load data
img=cv2.imread("AI.jpg")

##with the help of shape attribute, it is easy to find the image dimension 
if img is not None:
    height, width, channel = img.shape
    print(f"Image height: {height}\nImage width: {width} \nImage color channel: {channel}")
else:
    print("Error: Image not found.")
```
